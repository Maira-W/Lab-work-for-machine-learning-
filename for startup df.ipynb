```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
plt.ion()
```

```python
plt.scatter(df['R&D Spend'],df['Profit'],color = 'Blue', marker ='o')
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.title("R&D Spend vs Profit")
plt.show()
```

```python
x=df[["R&D Spend"]]
y=df["Profit"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
Saldf = LinearRegression()         
Saldf.fit(x_train, y_train)        
y_pred = Saldf.predict(x_test)  
```

```python
r2_score=r2_score(y_test,y_pred)
MSE=mean_squared_error(y_test,y_pred)
```

```python

X = df[["R&D Spend"]]
y = df["Profit"]
model = LinearRegression()
model.fit(X, y)

# Prediction function
def predict_profit(rnd_spend):
    profit = model.predict([[rnd_spend]])[0]
    return f"Predicted Profit: {profit:.2f}"

# Gradio interface
iface = gr.Interface(
    fn=predict_profit,
    inputs=gr.Number(label="R&D Spend"),
    outputs="text",
    title="Startup Profit Predictor (Gradio)",
    description="Enter R&D Spend to predict Profit"
)

iface.launch()
```
